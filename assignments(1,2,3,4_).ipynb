```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib  as mpl
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
#assignment 1 : question:1 
a=np.arange(40,50)
b=np.arange(50,60)
plt.plot(a,b)
plt.title("variation of a and b")
plt.xlabel("a values")
plt.ylabel("b values")
```

```python
#assignment 1 : question no:2
days=[1,2,3,4,5,6,7]
sales_1=[160,150,140,145,175,165,180]
sales_2=[70,90,160,150,140,145,175]
plt.plot(days,sales_1,label="days vs sales_1")
plt.plot(days,sales_2,label="days vs sales_2")
plt.title("days vs sales")
plt.xlabel("days of the week")
plt.ylabel("sales")
plt.legend()
plt.show()
```

```python
labels=['tatamotars','hyndai','honda','maruthi','toyato','nissan']
shares=[123,145,167,189,190,110]
colors=['orange','red','blue','green','yellow','brown']
plt.pie(shares,labels=labels,colors=colors,autopct="%1.2f%%",shadow=True)
plt.title("company shares")
plt.show()
```
